```python
import altair as alt
import pandas as pd
import geopandas as gpd
import json
```

```python
geo_json_file_loc= '../data/taxi_zones/taxi_zones.geojson'

with open(geo_json_file_loc) as json_data:
    data = json.load(json_data)
```

```python
df_all = gpd.GeoDataFrame.from_features((data))
df_man = df_all[df_all['borough']=='Manhattan']
df_man.head()
```

```python
df_man_tojson = df_man.loads()

selection = alt.selection_single();

mymap = alt.Chart(alt.Data(values=df_man_tojson['features'])).mark_geoshape(
    stroke='black'    
    ).encode(
    tooltip = [alt.Tooltip('zone:N'),
               alt.Tooltip('location_id:Q')]
    )

mymap
```
